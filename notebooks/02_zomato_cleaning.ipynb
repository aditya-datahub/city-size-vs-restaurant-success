```python
import pandas as pd
```

```python
zomato = pd.read_csv("zomato.csv", encoding="latin-1")
```

```python
zomato.head(2)
```

```python
zomato = zomato[[
    'Restaurant Name',
    'City',
    'Cuisines',
    'Average Cost for two',
    'Price range',
    'Aggregate rating',
    'Votes',
    'Has Online delivery',
    'Has Table booking'
]]
```

```python
zomato = zomato.rename(columns = {
    'Restaurant Name': 'restaurant_name',
    'City': 'city',
    'Cuisines': 'cuisines',
    'Average Cost for two': 'avg_cost_for_two',
    'Price range': 'price_range',
    'Aggregate rating': 'rating',
    'Votes': 'votes',
    'Has Online delivery': 'online_delivery',
    'Has Table booking': 'table_booking'
})
```

```python
def clean_city_name(city):
    if pd.isna(city):
        return city
    city = city.lower().strip()
    city = city.replace("bengaluru", "bangalore")
    city = city.replace("new delhi", "delhi")
    city = city.replace("gurugram", "gurgaon")
    city = city.replace("noida", "noida")  # keep same
    return city

zomato['city_clean'] = zomato['city'].apply(clean_city_name)
```

```python
zomato['rating'] = pd.to_numeric(zomato['rating'], errors='coerce')
zomato['votes'] = pd.to_numeric(zomato['votes'], errors='coerce')
```

```python
zomato = zomato.dropna(subset=['rating', 'votes'])
```

```python
# Remove rows with any missing value

zomato = zomato.dropna()
```

```python
# Save cleaned data

zomato.to_csv("zomato_cleaned.csv", index=False)
```
